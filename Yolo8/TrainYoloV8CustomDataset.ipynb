```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import OpenEXR
import Imath
from ultralytics import YOLO

# Load the trained YOLO model
model = YOLO('/content/drive/My Drive/Yolo8/runs/runs/detect/train/weights/best.pt')  # Update with the path to your trained model

# Function to run inference on a new image and find centroids
def run_inference_and_find_centroids(model, image_path):
    # Run inference
    results = model(image_path, save=True)

    # Extract bounding boxes
    boxes = results[0].boxes.xyxy  # x1, y1, x2, y2
    centroids = []
    for box in boxes:
        x1, y1, x2, y2 = box[:4]  # Extract x1, y1, x2, y2
        cx = (x1 + x2) / 2
        cy = (y1 + y2) / 2
        centroids.append((int(cx), int(cy)))

    return centroids

# Function to load depth from an EXR file
def load_exr_depth(exr_path, centroids, image_shape):
    file = OpenEXR.InputFile(exr_path)
    dw = file.header()['dataWindow']
    width, height = dw.max.x - dw.min.x + 1, dw.max.y - dw.min.y + 1

    # Load channels
    pt = Imath.PixelType(Imath.PixelType.FLOAT)
    (R, G, B) = [np.frombuffer(file.channel(c, pt), dtype=np.float32).reshape((height, width)) for c in "RGB"]
    depth_map = (R + G + B) / 3  # Assuming the depth information is stored in RGB channels equally

    # Scale centroids to match depth map size
    scale_x = width / image_shape[1]
    scale_y = height / image_shape[0]

    scaled_centroids = [(int(cx * scale_x), int(cy * scale_y)) for cx, cy in centroids]

    depths = []
    for cx, cy in scaled_centroids:
        if 0 <= cy < height and 0 <= cx < width:
            depth = depth_map[cy, cx]
            depths.append(depth)
        else:
            depths.append(0.0)  # Default depth if centroid is outside image bounds

    return depths

# Function to display image with depth information at centroids
def display_image_with_depth(image_path, centroids, depths):
    # Load the image
    image = cv2.imread(image_path)

    # Overlay depth information on the image
    for (cx, cy), depth in zip(centroids, depths):
        label = f'Depth: {depth:.2f}m'
        cv2.putText(image, label, (cx, cy), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 1)
        cv2.circle(image, (cx, cy), 5, (0, 255, 0), -1)

    # Display the image
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.imshow(image)
    plt.axis('off')
    plt.show()

# Paths to the new image and corresponding EXR file for prediction
image_path = '/content/drive/My Drive/Yolo8/test/00031Right.png'  # Update with your image path
exr_path = '/content/drive/My Drive/Yolo8/depth_exr/00031Right.exr'    # Update with your EXR file path

# Run inference and find centroids
centroids = run_inference_and_find_centroids(model, image_path)

# Load depth map from EXR file
depths = load_exr_depth(exr_path, centroids, cv2.imread(image_path).shape)

# Display the image with depth information at centroids
display_image_with_depth(image_path, centroids, depths)
```
